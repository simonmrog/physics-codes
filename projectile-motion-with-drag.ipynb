```python
import math
import matplotlib.pyplot as plt
```

```python
# Initial values and lists
# The drag force is of the form F = -mkv
# The equations of motion are solved mathematically and it's result put directly into the code

v0 = 10.0

theta = math.radians(45);
v0x = v0 * math.cos(theta) # equations of motion
v0y = v0 * math.sin(theta)

g = 9.8
dt = 0.001

x = list() # array for x-position
y = list() # array for y-position

for k in range (0, 6): #calculation for k = 0, 1, 2, 3, 4, 5
    
    xi = 0.0
    yi = 0.0
    x.clear()
    y.clear()
    t = 0.0
    flag = 0
    
    while yi >= 0.0:
            
        if k == 0:
            xi = v0x * t
            yi = v0y * t - 0.5 * g * t * t
            
        else:
            xi = (v0x / k) * (1.0 - math.exp(-k * t))
            yi = (1.0 / k) * (v0y + g / k) * (1.0 - math.exp(-k * t)) - (g / k) * t
            
        t = t + dt
            
        x.append(xi)
        y.append(yi)
            
    plt.plot(x, y)
    
plt.title("Projectile motion for different values of k")
plt.xlabel("x (metres)")
plt.ylabel ("y (metres)")
plt.show()
```
