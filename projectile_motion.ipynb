```python
# Modules to import

import matplotlib.pyplot as plt
import math
```

```python
# Initial values and lists

xi = 0
yi = 0
v0 = 10
t0 = 0

g = 9.8
dt = 0.001

x = list() # array for x-position
y = list() # array for y-position

for theta in range (5, 85, 5): # sweeps angles from 5 to 85
        
    theta_rad = math.radians(theta) # angle conversion
    v0x = v0 * math.cos(theta_rad) # equations of motion
    v0y = v0 * math.sin(theta_rad)
    t = t0
        
    x.clear()
    y.clear()
    tf = 2 * v0 * math.sin(theta_rad) / g # flight time for every case
    
    while t <= tf:
            
        xi = v0x * (t - t0)
        yi = v0y * (t - t0) - 0.5 * g * (t - t0) * (t - t0)
        t = t + dt
            
        x.append(xi)
        y.append(yi)
            
    plt.plot(x, y)
    
plt.title("Projectile motion for different angles")
plt.xlabel("x (metres)")
plt.ylabel ("y (metres)")
plt.show()
```
